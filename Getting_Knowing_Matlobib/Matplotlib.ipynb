# Pyplot API vs Object-Oriented API  
### Pyplot API -> nhanh hơn 
### OO API -> vẽ đồ thị advanced hơn 

## Pyplot API

```python
x = np.linspace(0,10,1000) 
plt.plot(x, np.sin(x), color="blue", label="sin(x)"); #đồ thi hình sin 

#Vẽ đồ thị khác chung 1 hình 
plt.plot(x,np.cos(x), color="red", label ="cos(x)")

#Tên trục x,y
plt.xlabel("Biến x")
plt.ylabel("Sin(x)") 

#Giới hạn cho trục x,y 
plt.xlim([0,4]) 
plt.ylim([-0.7,0.75]) 

#show ra label của các đường trong đồ thị 
plt.legend(); 

# show ra tên của đồ thị
plt.title("Sin(x) and Cos(x)")

# plt.axis([xmin, xmax, ymin, ymax]) --> tương tự như xlim và ylim 
plt.axis([0,4,-0.7,0.75])

plt.axis("equal") #Làm cho trục x và y bằng nhau 
```

## Object-Oriented API

```python
# Prepare data 
x = [1,2,3,4] 
y = [11,22,33,44]

# setup plot 
fig, ax = plt.subplots(figsize=(5,5)) #Thiết lập size với width và height của đồ thị

# plt.subplots() là một hàm của Matplotlib dùng để tạo một figure (hình) và một hoặc nhiều trục (axes).
# fig là đối tượng đại diện cho toàn bộ "hình vẽ" (figure), nó bao gồm tất cả các thành phần của đồ thị.
# ax là đối tượng "trục tọa độ" (axes), nơi dữ liệu được biểu diễn. Nó là nơi thực tế mà các đường đồ thị được vẽ.

# Plot the data
ax.plot(x,y) ; # ax.plot() là một phương thức của đối tượng ax dùng để vẽ đồ thị. Nó vẽ các điểm (x, y) và kết nối chúng bằng một đường.

# Customize the plot 
ax.set(title = "Đồ thị", xlabel = "Trục x", ylabel ="Trục y");
```

# Các loại đồ thị phổ biến trong Matplotib

## Line chart

```python
# Create array 
x = np.linspace(0,10,100)
```

```python
# )) API plot a line chart 
fig, ax = plt.subplots()
ax.plot(x, x**3); #Hàm số có dạng y = x ^ 3
```

## Scatter 

```python
plt.scatter(x, np.exp(x)); # y = e^x 
```

```python
# OO API 
# Prepare new data 
rng = np.random.RandomState(0) #Tạo các giá trị random vs seed cố định là 0 -> đảm bảo giá trị ngẫu nhiên luôn giống nhau khi chạy lại
x = rng.randn(100) # Sinh ra 100 giá trị ngẫu nhiên theo phân phối chuẩn với mean = 0, độ lệch chuẩn = 1, và đc gán cho biến x
y = rng.randn(100)

sizes = 1000*rng.rand(100) # Tạo 1 mảng gồm 1000 phần tử với 100 giá trị ngẫu nhiên từ 0 đến 1
colors = 1000*rng.rand(100)
fig,ax = plt.subplots()
img1 = ax.scatter(x,y, s=sizes, c=colors, cmap='viridis', alpha = 0.7 ); # Áp dụng bảng màu (color map) 'viridis' và độ trong suốt là 0.3
fig.colorbar(img1);
```

## Bar chart

### Vertical bar chart

```python
# Prepare data 
soft_drink_prices = {"Coke":10, 
                    "Pepsi":12, 
                    "Sprite":8}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values()); # truc x là key, trục y là value
ax.set(title="Coopmart đồ uống", ylabel ="nghìn VNĐ"); 
```

### Horizontal bar chart 

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values())); #Do hàm barh k nhận dictionary là giá trị đầu vào nên phải chuyển sang list
```

## Histogram: Biểu đồ thể hiện tần suất dạng cột 

```python
# Prepare data 
np.random.seed(42)
student_height = np.random.normal(170,10,250) # Giá trị tb là 170, độ lệch chuẩn 10
```

```python
# Pyplot API 
plt.hist(student_height, bins = 30); #Chia nhỏ ra
```

## Subplots : cách vẽ subplots dùng OO API

### Cách 1 

```python

fig,((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))

#Plot data to each axis 
ax1.plot(x,x/2); 
ax1.plot(x,x*2);

ax2.scatter(np.random.random(10),np.random.random(10));

ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());

ax4.hist(student_height, bins = 30);
```

### Cách 2: 

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))

ax[0,0].plot(x,x/2); 
ax[0,0].plot(x,x*2);

ax[0,1].scatter(np.random.random(10),np.random.random(10));

ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());

ax[1,1].hist(student_height, bins = 30);
```

# Small project : Data visualization : California Cities 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
cities = pd.read_csv('california_cities.csv')
cities.head()
```

```python
# Extract Latd ( vĩ độ ) and Longd (Kinh độ ) 
lat, long = cities['latd'], cities['longd']
```

```python
lat[:10]
```

```python
# Plot using Pyplot API 
population, area = cities['population_total'], cities['area_total_km2']
plt.style.use('classic')
plt.figure(figsize =(8,6))
plt.scatter(long,lat, c = np.log10(population), s = area, linewidths = 0, alpha = 0.5 ) 
#do hầu hết các thành phố đều đông dân nên ta dùng log10 để số bé lại -> màu sắc hơn
plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)'); #Thanh hiển thị màu ( nhớ để đơn vị thành log10)

# Create a legend for cities size 
area_range = [50,100,300,500] 
for area in area_range: 
    plt.scatter([],[], s = area, c='red', label = str(area) + 'km$^2$')
plt.legend(scatterpoints = 1, labelspacing = 0.5, title='City Area')


plt.clim(3,7) ; # thanh color từ 3-> 7
plt.title('Califorina Cities: Population and Area Distribution');
```
